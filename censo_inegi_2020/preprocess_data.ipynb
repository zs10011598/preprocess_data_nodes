```python
df = pd.read_csv('state_01/conjunto_de_datos/conjunto_de_datos_ageb_urbana_01_cpv2020.csv')
df = df.iloc[0:0]

for i in range(1, 33):
    state_code = str(i).zfill(2)
    df_aux = pd.read_csv('state_{state_code}/conjunto_de_datos/conjunto_de_datos_ageb_urbana_{state_code}_cpv2020.csv'.format(state_code=state_code))
    df_aux = df_aux[(df_aux['MUN'] == 0) & (df_aux['AGEB'] == '0000') & (df_aux['MZA'] == 0)]
    
    try:
        df = pd.concat([df, df_aux])
    except Exception as e:
        print(i, str(e))
        
df = df.reset_index()
df = df.drop(columns=['index', 'NOM_ENT', 'MUN', 'NOM_MUN', 'LOC', 'NOM_LOC', 'AGEB', 'MZA'])
df['ENTIDAD'] = df['ENTIDAD'].apply(lambda x: str(x).zfill(2))
df.to_csv('./states.csv', index=False)


df
```
